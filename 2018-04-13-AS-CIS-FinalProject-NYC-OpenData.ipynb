```python
import pandas as pd
from IPython.display import display
import numpy as np
```

```python
data =pd.read_table("weatherData.csv", sep=",")
pd.options.display.max_columns = None
display(data)
```

```python
data['RainStatus']= np.where(data["PRCP"]>0,'1','0')
data['SnowStatus']= np.where(data["SNOW"]>0,'1','0')
```

```python
data.loc[:,['PRCP','RainStatus','SNOW','SnowStatus']]
```

```python
data['weatherType']= np.where(data["PRCP" and "SNOW"]>0,'1','0')
```

```python
data.loc[:,['PRCP','RainStatus','SNOW','SnowStatus','weatherType']]
```

```python
data.to_csv('CleanWeatherData.csv')
```
